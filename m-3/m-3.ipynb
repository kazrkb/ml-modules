```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    {'h': [150,160,170,180,190],
    'w': [53,67,71,80,90]}
)
df
```

```python
df.describe()
```

```python
z_scaling = round((df - df.mean())/df.std(),2)
z_scaling
```

```python
z_scaling.plot(kind='bar')
plt.title("Z Scaling")
plt.xlabel("Index"); plt.ylabel("z value")
plt.show()
```

```python
df
```

```python
min_max_scaling = (df - df.min())/(df.max()-df.min())
min_max_scaling
```

```python
min_max_scaling.plot(kind='bar')
plt.title("min max Scaling")
plt.xlabel("Index"); plt.ylabel("min max value")
plt.show()
```
