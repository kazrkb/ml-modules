```julia
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```julia
df = pd.DataFrame(
    {'h': [150,160,170,180,190],
    'w': [53,67,71,80,90]}
)
df
```

```julia
df.describe()
```

```julia
z_scaling = round((df - df.mean())/df.std(),2)
z_scaling
```

```julia
z_scaling.plot(kind='bar')
plt.title("Z Scaling")
plt.xlabel("Index"); plt.ylabel("z value")
plt.show()
```

```julia
df
```

```julia
min_max_scaling = (df - df.min())/(df.max()-df.min())
min_max_scaling
```

```julia
min_max_scaling.plot(kind='bar')
plt.title("min max Scaling")
plt.xlabel("Index"); plt.ylabel("min max value")
plt.show()
```

```julia
# robust scaling
df
```

```julia
iqr = df.quantile(0.75) - df.quantile(0.25)
robust = (df - df.mean())/iqr
robust
```

```julia
robust.plot(kind='bar')
plt.title("min max Scaling")
plt.xlabel("Index"); plt.ylabel("robust value")
plt.show()
```

```julia
import pandas as pd

# 1. Create the data for the dataset
data = {
    'id': [1, 2, 3, 4],
    'Age': [25, 42, 22, 35],
    'City': ['New York', 'London', 'Paris', 'Tokyo'],
    'Level': ['Mid', 'Senior', 'Junior', 'Mid'],
    'Score': [85.5, 92.0, 78.5, 88.0]
}

# 2. Create the Pandas DataFrame
df = pd.DataFrame(data)

# 3. Print the DataFrame
print(df)
```

```julia
# one hot encoding

d_city = pd.get_dummies(df['City'], prefix='C', dtype=int)

df_encoded = pd.concat([df, d_city],axis=1)
df = df_encoded.drop('City', axis=1)
print(df)
```
