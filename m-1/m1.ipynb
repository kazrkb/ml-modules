```python
import numpy as np
import pandas as pd

data = [120,125,128,122,126,124,600]  #600 is an outliers
df = pd.DataFrame({'HousePrice': data})

mean_value = df['HousePrice'].mean()
print(f"Mean: {mean_value}")
```

```python
median_value = df['HousePrice'].median()
print(f"Median: {median_value}")
```

```python
# categorical data 

df['City'] = ['Dhaka','Dhaka','Chottogram','Sylhet','Dhaka','Chottogram','Chottogram', ]
mode_value = df['City'].mode()[0]
print(f"Mode: {mode_value}")
```

```python
from sklearn.impute import SimpleImputer

#mean good for balanced data when there is no outliers
imputer_mean = SimpleImputer(strategy='mean')

# good for skewed or outliers rich data
imputer_median = SimpleImputer(strategy='median')

# best for categorical data
impute_mode = SimpleImputer(strategy='most_prequent')
```
